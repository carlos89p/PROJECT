## model train

```python
from ultralytics import YOLO
from pathlib import Path

root = Path().resolve()
yaml_path = root / "yolo_dataset" / "data.yaml"

model = YOLO("yolov8n.pt")  # modelo mÃ¡s rÃ¡pido

model.train(
    data=str(yaml_path),
    epochs=30,                # solo 5 para pruebas
    imgsz=416,               # tamaÃ±o mÃ¡s pequeÃ±o
    batch=8,                 # menos carga
    name="fast_test_model",
    project="runs/train",
    pretrained=True
)
```

```python
from ultralytics import YOLO
import cv2

# Cargar el modelo entrenado
model = YOLO("runs/train/fast_test_model/weights/best.pt")

# Ruta al vÃ­deo de entrada
video_path = "traffic_sequence.mp4"  # Cambia esto si tu archivo tiene otro nombre

# Captura de vÃ­deo
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("âŒ No se pudo abrir el vÃ­deo.")
else:
    # Obtener info del vÃ­deo
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)

    # (Opcional) Guardar salida
    out = cv2.VideoWriter("output_detected.mp4", cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

    # Clases que tu modelo conoce
    class_names = ["red", "yellow", "green"]

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Ejecutar detecciÃ³n
        results = model(frame, imgsz=416, conf=0.1, verbose=False)[0]

        # Dibujar resultados
        for box in results.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cls_id = int(box.cls[0])
            conf = float(box.conf[0])
            label = f"{class_names[cls_id]} {conf:.2f}"

            color = {
                "red": (0, 0, 255),
                "yellow": (0, 255, 255),
                "green": (0, 255, 0)
            }[class_names[cls_id]]

            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

        # Mostrar en tiempo real
        cv2.imshow("DetecciÃ³n de semÃ¡foros", frame)
        out.write(frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    out.release()
    cv2.destroyAllWindows()
    print("âœ… Proceso terminado. VÃ­deo guardado como 'output_detected.mp4'")
```

```python
import cv2
import os
import matplotlib.pyplot as plt

# Cambia esto al nombre de una imagen real que exista
image_name = "00065.jpg"

# Rutas
image_path = f"yolo_dataset/images/train/{image_name}"
label_path = f"yolo_dataset/labels/train/{image_name.replace('.jpg', '.txt')}"

# Cargar imagen
img = cv2.imread(image_path)
h, w, _ = img.shape

# Cargar etiquetas
if os.path.exists(label_path):
    with open(label_path, "r") as f:
        lines = f.readlines()
        for line in lines:
            parts = line.strip().split()
            cls = int(parts[0])
            x_c, y_c, bw, bh = map(float, parts[1:])
            # Convertir a coordenadas absolutas
            x1 = int((x_c - bw / 2) * w)
            y1 = int((y_c - bh / 2) * h)
            x2 = int((x_c + bw / 2) * w)
            y2 = int((y_c + bh / 2) * h)
            # Dibujar caja
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(img, f"Class {cls}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)
else:
    print("âŒ No se encontrÃ³ el archivo de etiquetas.")

# Mostrar imagen
plt.figure(figsize=(10, 8))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("VisualizaciÃ³n de etiquetas YOLO")
plt.axis("off")
plt.show()
```
