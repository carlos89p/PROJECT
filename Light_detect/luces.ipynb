## entrenamos el modelo Yolo

```python
from ultralytics import YOLO
from pathlib import Path

# === Ruta absoluta al archivo YAML ===
root = Path().resolve()
yaml_path = root / "lights" / "data.yaml"

# === Cargar modelo mÃ¡s preciso (YOLOv8s) ===
model = YOLO("yolov8s.pt")

# === Entrenamiento ===
model.train(
    data=str(yaml_path),           
    epochs=30,                    
    imgsz=416,                     
    batch=16,                      
    name="traffic_light_detector", 
    project="runs/train",
    pretrained=True,              # Fine-tuning desde COCO
    patience=10,                  # Detiene si no mejora tras 10 Ã©pocas
    verbose=True,                 # Ver mÃ¡s detalles en consola
    plots=True                    # Guarda grÃ¡ficas como results.png
)
```

## Deteccion de semaforos en imagenes random

```python
%matplotlib inline

from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt
from pathlib import Path
import random
import os

# === CARGAR MODELO DE SEMÃFOROS ===
model = YOLO("runs/train/traffic_light_detector/weights/best.pt")

# === DIRECTORIO CON IMÃGENES DE TEST ===
image_dir = Path("lights/test/images")

# === COMPROBAR QUE EXISTE ===
if not image_dir.exists():
    raise FileNotFoundError(f"No se encontrÃ³ la carpeta: {image_dir}")

# === SELECCIONAR UNA IMAGEN ALEATORIA ===
image_list = list(image_dir.glob("*.jpg"))
if not image_list:
    raise FileNotFoundError("No se encontraron imÃ¡genes .jpg en el directorio.")

image_path = random.choice(image_list)

# === CARGAR LA IMAGEN ===
img = cv2.imread(str(image_path))
if img is None:
    raise ValueError(f"No se pudo cargar la imagen: {image_path}")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# === INFERENCIA ===
results = model(img_rgb, conf=0.3)[0]

# === DIBUJAR RESULTADOS ===
for box in results.boxes:
    x1, y1, x2, y2 = map(int, box.xyxy[0])
    conf = float(box.conf[0])
    label = model.names[int(box.cls[0])]
    color = (0, 255, 0)
    cv2.rectangle(img_rgb, (x1, y1), (x2, y2), color, 2)
    cv2.putText(img_rgb, f"{label} {conf:.2f}", (x1, y1 - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

# === MOSTRAR LA IMAGEN DETECTADA ===
plt.figure(figsize=(10, 8))
plt.imshow(img_rgb)
plt.title(f"DetecciÃ³n de semÃ¡foros en: {image_path.name}")
plt.axis("off")
plt.show()
```

## indicaciones de que hacer en funcion del semaforo

```python
%matplotlib inline

from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt
from pathlib import Path
import random
import os

# === CARGAR MODELO ENTRENADO ===
model = YOLO("runs/train/traffic_light_detector/weights/best.pt")

# === DIRECTORIO CON IMÃGENES DE TEST ===
image_dir = Path("lights/test/images")

# === COMPROBAR QUE EXISTE ===
if not image_dir.exists():
    raise FileNotFoundError(f"No se encontrÃ³ la carpeta: {image_dir}")

# === SELECCIONAR UNA IMAGEN ALEATORIA ===
image_list = list(image_dir.glob("*.jpg"))
if not image_list:
    raise FileNotFoundError("No se encontraron imÃ¡genes .jpg en el directorio.")

image_path = random.choice(image_list)

# === CARGAR LA IMAGEN ===
img = cv2.imread(str(image_path))
if img is None:
    raise ValueError(f"No se pudo cargar la imagen: {image_path}")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# === INFERENCIA ===
results = model(img_rgb, conf=0.3)[0]

# === DIBUJAR RESULTADOS + DETERMINAR LA LUZ CON MAYOR ÃREA ===
largest_area = 0
status = None

for box in results.boxes:
    x1, y1, x2, y2 = map(int, box.xyxy[0])
    area = (x2 - x1) * (y2 - y1)
    conf = float(box.conf[0])
    cls = int(box.cls[0])
    label = model.names[cls].lower()

    # Dibujar todas las cajas
    cv2.rectangle(img_rgb, (x1, y1), (x2, y2), (0, 255, 0), 2)
    cv2.putText(img_rgb, f"{label} {conf:.2f}", (x1, y1 - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Guardar la luz mÃ¡s grande
    if area > largest_area:
        largest_area = area
        if label == "merah":
            status = "STOP"
        elif label == "kuning":
            status = "CAUTION"
        elif label == "hijau":
            status = "CONTINUE"

# === MOSTRAR ESTADO EN RECUADRO NEGRO CON TEXTO ROJO ===
if status:
    h, w, _ = img_rgb.shape
    cv2.rectangle(img_rgb, (w - 220, h - 70), (w - 20, h - 20), (0, 0, 0), -1)  # fondo negro
    cv2.putText(img_rgb, status, (w - 210, h - 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3)  # texto rojo

# === MOSTRAR RESULTADO ===
plt.figure(figsize=(10, 8))
plt.imshow(img_rgb)
plt.title(f"DetecciÃ³n de semÃ¡foros en: {image_path.name}")
plt.axis("off")
plt.show()
```
